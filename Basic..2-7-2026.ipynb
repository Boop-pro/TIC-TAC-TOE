```python
#                        TIC-TAC-TOE
import random
def get_choice():
    player_choice = input("Enter a choice(rock,paper,scissors):")
    options = ["rock","paper","scissors"]
    computer_choice = random.choice(options)
    choices = {"player":player_choice,"computer":computer_choice}
    return choices

def check_win(player,computer):
    print(f"You choose {player} ,computer choose {computer}")
    if player == computer:
        print("."*40)
        return "tie"
    elif computer =="rock" and player == "paper" or computer == "scissors" and player == "rock" or computer == "paper" and player == "scissors":
        print("."*40)
        return "player"
    elif computer == "paper" and player == "rock" or computer =="rock" and player == "scissors" or computer == "scissors" and player == "paper":
        print("."*40)
        return "computer"
   
yo = get_choice()
print("The winner is :\t" + check_win(yo["player"],yo["computer"]))
```

```python
#                        Wage Calculator
start_hour = int(input("Enter the start hour (0-23): "))
start_minute = int(input("Enter the start minute (0-59): "))
end_hour = int(input("Enter the end hour (0-23): "))
end_minute = int(input("Enter the end minute (0-59): "))
hourly_wage = float(input("Enter the hourly wage: "))
print(f"Worked: {start_hour}:{start_minute:02d} to  {end_hour}:{end_minute:02d}")

worked_hours = ((end_hour - start_hour)) + ((end_minute - start_minute)) / 60
total_earnings = worked_hours * hourly_wage
print(f"Total hours worked: {worked_hours:.2f}")
print(f"Total earnings: ${total_earnings:.2f}")
```
